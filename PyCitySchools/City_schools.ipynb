```python
num_schools=school_data_complete["school_name"].nunique()
num_students=school_data_complete["Student ID"].count()
school_budget=school_data["budget"].sum()
math_score_mean=school_data_complete["math_score"].mean()
reading_score_mean=school_data_complete["reading_score"].mean()
pass_math=school_data_complete["math_score"].loc[school_data_complete["math_score"]>=70].count()
pass_reading=school_data_complete["reading_score"].loc[school_data_complete["reading_score"]>=70].count()
pass_math_average=pass_math/num_students*100
pass_reading_average=pass_reading/num_students*100
pass_overall=(math_score_mean+reading_score_mean)/2


school_summary=pd.DataFrame([{"Total Schools":num_schools,"Total Students":num_students,"Total Budget":school_budget\
              ,"Average Math Score":math_score_mean,"Average Reading Score":reading_score_mean,\
              "% Passing Math":pass_math_average,"% Passing Reading":pass_reading_average,\
               "% Overall Passing Rate":pass_overall}])
school_summary.head()
```
